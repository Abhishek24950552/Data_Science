## Pandas

```python
pandas >> Python Data Analysis
Pandas Library is built on topof the Numpy Library

Two data structure in pandas:
    1. DataFrame --> 2-D Data Structure contains Rows and Columns
    2. Series--> One Dimensional Data Structure

Use cases of pandas:
    1. Read Data From CSV, Excel, JSON
    2. Read Data From Database(SQL or NoSQL)
    3. Exploratory Data Analysis
    4. Data Cleansing:
        Missing Values
        Handle Outliers    
    5. Encoding:
        1. Label Encoding
        2. One Hot Encoding    
    6. Feature Selection
    7. Join Multiple Tables
    8. Provides Time-Series Functionality
```

## Pandas Functions Overview

```python
df.columns--> used to get list of column names

df.index --> used to get list of row lables

df.axes --> gives list of column names and row lables

df.dtypes --> list out data types of all columns

df.select_dtypes(include, exclude) --> is used to include or exclude columns with perticular dtypes

df.info() --> is used to get info like null values, names of column, dtypes of columns

df.describe() --> it is used to get  max, std, count, min, mean, median, 50% 25% 75% values

df.shape--> it is used to get shape of dataframe df.shape[0]-->rows df.shape[1]-->colomns

Access Rows and Columns:
-->we can access columns directly by using column names
    df[column name /list of column names]
    df[Col_name]
    df[[Col1, Col2,... ,ColN]]

df.iloc: it is used to get values by using index of column and index of row
    -->here index of row is compulsory and column index is optional
    -->we can also acces perticular cell values
    -->df.iloc[1,0]-->1 index row and 0 index col cell value and o/p will be series
        df.iloc[[1],[0]]-->1 index row and 0 index col cell value and o/p will be dataframe
    -->it works as slicing end index value is excluded 
    1. df.iloc[row_index, col_index]
    2. df.iloc[row_index]
    3. df.iloc[:,col_index]
    4. df.iloc[row_index_list, col_index_list]
    
df.loc:
    -->we use lable names to access values 
    -->start and end labels will be included
    1. df.loc[row_labels, col_labels]
    2. df.loc[row_labels]
    3. df.loc[:,col_labels]
             
df[start_row_index : end_row_index]--> Acces Rows 

df.head()-->First N rows of dataframe, Default Value is 5

df.tail()-->Last N rows of dataframe, Default Value is 5 


Iterate DataFrame using For Loop:
    1. df.items()-->Iterates over the DataFrame columns, returning a the column name and the content as a Series.
    2. df.iteritems()-->use is same as df.items()
    3. df.iterrows()-->Iterate over DataFrame rows as (index, Series) pairs
    
DataFrame to Numpy:
    1. df.values-->Return a Numpy representation of the DataFrame. use to_numpy()(recommeded in pandas doc)
    2. df.to_numpy()-->Convert the DataFrame to a NumPy array.
    3. df.values.tolist()-->Convert numpy array to list
    4. df.to_numpy().tolist()-->Convert numpy array to list

Detect Missing Values:
    1. df.isna()-->Returns dataframe with true and false values. only True for null values in dataframe
    2. df.isnull()-->same as isna()
    3. df.isna().sum()-->Gives count of null values in each column
    4. df.isnull().sum()-->same as df.isna().sum()
    5. df.isna().mean()*100-->gives % of null values in each column 
    
Drop Null Values:
--> it used to drop rows and columns with specific null values
    syntax--> df1.dropna(axis,thresh)
    axis-->0 row operation
    axis-->1 col operation
    thresh--> acceptable null values in row or column   
    
    1. df.dropna(thresh = col_count, axis = 0, inplace = False)
       df.dropna(thresh = row_count, axis = 1, inplace = False)    
    
--> df.drop(list of column names/list of row lables,axis,inplace)
    we need to pass list of column or row names
    default axis is 0,  0-->row action 1-->column action
    inplace-->default False,   True--> make changes in df False-->no changes in original df
    
    1. df.drop('Col_Name', axis = 1, inplace = False)
    2. df.drop([Col1, Col2,...,ColN], axis = 1, inplace = True/False)
    3. df.drop("Row Name",axis = 0, inplace = False)
       df.drop("Row Name", inplace = False)
    4. df.drop([Row1,Row2...,RowN],axis = 0)
    
    
Impute Missing Values:
    It is used to fill null values with given values or stats values in column
    
    1. df.fillna(value, inplace = False)-->fills all null values in dataframe
    2. emp_df['column'].fillna(emp_df['column'].stat_fun(),inplace=True/False)-->fill null values of specific column
    
    Statistical Methods:
        These methods can be used to fill values in perticular column
        1. df['ColName'].mean()
        2. df['ColName'].median()
        3. df['ColName'].mode()[0]
        
        Mean >> If outliers are not present in data
        Median >> If outliers are present in data
        Mode >> Categorical Data


df['Col_Name'].unique()-->it is used to get unique values in column
df['Col_Name'].nunique()-->it gives count of unique values in column
df['Col_Name'].value_counts()-->it gives unique values and their count in column

pd.crosstab(x = df['Col_Name1'], y = df['Col_Name2'])-->Categorical variables

Series to Dict:
    pd.Series.to_dict()
    df['Col_Name'].value_counts().to_dict()
    
df['Col_Name'].astype(datatype) --> casting dtype

df['Col_Name'].replace({dict})-->it is used to replace values in column, we pass values in dict form

sns.countplot(x = df['Col_Name1'], hue = df['Col_Name2'])

pd.sort_index(axis=0)-->it is used to sort index(rows) 

pd.sort_index(axis=1)-->it is used to column lables(colums)

pd.sort_values(column name,ascending=True)-->sort values of perticular column 

df1.to_csv('file name')-->convert dataframe to csv file

pd.sort_values([column1,column2],ascending=True)-->sort values based on multiple columns 

df.reset_index(drop=True)-->drops old index column and creates new 

df.set_index(col name)-->it is used to set perticular column as index

df.set_index(column name).sort_index()-->set column as index and then sort that index

df.rename({'original':'new name'},axis=1/0,inplace=True/False)-->used to rename label names

apply():
    Apply a function along an axis of the DataFrame.
    Objects passed to the function are Series objects whose index is either the DataFrame’s 
    index (axis=0) or the DataFrame’s columns (axis=1). 
    syntax:
        pd.apply(function/lambda)

map():
    map is used to make changes but the cell where condition is not satisfied are made null
    syntax:
        pd.map(dict)

replace():
    replace is used to replace values but only changes are made in cell where condition is satisfied 
    and all other cell are same 
    syntax:
        pd.replace(dict)

groupby():
    Group DataFrame using a mapper or by a Series of columns.   
    # to get group and only index of row
    mov.groupby('genre').groups 

    # get only keys(unique values in column) same as unique function 
    mov.groupby('genre').groups.keys() 
    
    # group by two columns and getting records from that values
    mov.groupby(['col1','col2',..]).get_group(('group value1','group value2',..))
    
pd.loc():
    # accessing using loc function note [] brackets
    mov.loc[(mov['column']=='value')]

    # .isin() function to get multiple vales
    # we need to pass list of items and use logical operators
    mov.loc[(mov['column1'].isin(['val1', 'val2', 'val3'])&(mov['column2'].isin(['val1', 'val2'])))]

get_dummies():
    to perform one hot encoding we are using get_dummies() function
    citpd.get_dummies(df['colum'],prefix='column prefix')

pd.drop_duplicates():
    it is used to delete duplicate rows in dataframe
    syntax:
    df.drop_duplicates()-->drops duplicate rows
    df.drop_duplicates(keep='first')-->keeps first value and deletes other values
    df.drop_duplicates(keep='last')-->keep last values and deletes other values

pd.read_csv():
    syntax:
    pd.read_csv(file_path,usecols=[list of columns],header=None,index_col='column name',delimiter=',|')
    path-> file path
    usecols-->columns to display
    header-->none
    index_col-->make column as index
    delimeter-->delimiter for seperation

pd.read_excel()
    syntax:
    df = pd.read_excel(excel_file_path, sheet_name='Sheet_Name', usecols=[columns_list], 
                index_col='Col_Name',
                header=None)
    
to_csv():
    to create csv files
    df.to_csv(csv_file_path, index = True, header= True)

Create Excel File:
    1. Create single Sheet Excel File:
        df.to_excel('Excel_File_Path.xlsx', index = True, header=True, sheet_name = 'SheetName')
    
    2. Create Multiple Sheets in Excel File:
        1. Writer = pd.ExcelWriter("Excel_File_Path.xlsx",engine='xlsxwriter')
        2. df1.to_excel(writer, index=True, header=True, sheet_name = "Sheet1")
        3. df2.to_excel(writer, index=True, header=True, sheet_name = "Sheet2")
        4. df3.to_excel(writer, index=True, header=True, sheet_name = "Sheet3")
        5. Writer.save()

creating dict:
    data1.to_dict() # here default orient is dict and create dict of dict
    data1.to_dict(orient='list') # here orient is list and will create dict of list
    data1.to_dict(orient='records') # here orient is records and will print row wise data
    
creating json files:
    new_df.to_json('employee_data.json')
    new_df.to_json('employee_data2.json',orient='records')
    
insert():
    syntax:
        df.insert(index, Col_Name, value, allow_duplicates=True/False)

append():
    syntax:
    df1.append(df2,ignore_index=False/True)

concat():    
    syntax:
        new_df = pd.concat([df1,df2,...],ignore_index=False,axis = 1,join='outer'/'inner')
    here operations are performed on column and row names
    by default it performs outer join
    all unique values are present in result
    ignore_index-->to ignore_index from original dataframes(True/False) it removes row/column names based on axis
    axis--> 1=column addidition operation 0=row addidition operation
    join--> 'outer'-->it is by default and gives all unique values
            'inner'-->it gives only common index and rows based on axis
        
merge():
    syntax:
    pd.merge(df1,df2,on=column name which is common,how=inner/outer/left/right,suffixes=['_suffix1','_suffix2'])
    merge works on values of dataframe
    how-->which join we want default is inner
    on-->column on which we compare values
    left_on and right_on for diff column names 
    
join():
    syntax:
    df1.join(df2,how = 'left/right/inner/outer',lsuffix='_df1',rsuffix='_df2')
    
datetime():
    it is used to convert obj type to datetime 
    syntax:
    df[column] = pd.to_datetime(df[column],format = 'date format')
    date format should be same as format in column
    
aggregate() and agg():
    df.aggregate(['sum','min','max','mean','std','var'])
    df.agg(['sum','min','max','mean','std','var'])
```

## pd.apply()

```python
Apply a function along an axis of the DataFrame.
Objects passed to the function are Series objects whose index is either the DataFrame’s 
index (axis=0) or the DataFrame’s columns (axis=1). 
```

## map() and replace()

```python
map():
    map is used to make changes but the cell where condition is not satisfied are made null

replace():
    replace is used to replace values but only changes are made in cell where condition is satisfied 
    and all other cell are same 
```

```python
m = pd.read_csv('movies_data.csv')
mov =m.head(5)
print(mov)

# map() will makes changes to given values and other values will be null
print('\n this is result for map function \n')
print(mov['genre'].map({'Action' : "ACTION"}))
print(mov['genre'].map({'Crime':'CRIME','Action' : "ACTION"}))

# replace() will makes changes in given values and other values will remain same
print('\n this is result for replace function \n')
print(mov['genre'].replace({'Crime':'CRIME'}))
```

## Encoding


```python
Data:
    Categorical Data
    Continuous Data
    
Categorical Data:
    1. Ordinal Data: Data will be in order
        1. High - Medium - Low
    2. Nominal Data : Data without order or Rank
        
        
Ordinal Data >> Label Encoding
Nominal Data >> One Hot Encoding
```

### label encoding

```python
data = {'Education' : ['SSC', 'PG', 'Grad','PG','HSC','Phd','PG','Grad','HSC','Grad'],
       'City_Names' : ['Pune','Mumbai','Delhi','Pune','Pune','Chennai','Mumbai','Delhi',
                      'Pune','Chennai'],
       "Genre" : ['Crime', 'Action', 'Drama', 'Western', 'Adventure',
                 'Action', 'Drama', 'Western','Crime', 'Action'],
       'Duration' : [142, 175, 200, 152, 154,  96, 161, 201, 195,104]}

df = pd.DataFrame(data)
df
# to perform label encoding

# first get unique values from from row and assign them values for encoding
# value_counts() gives unique values and its counts
# to_dict() function is used to get dict
df.Education.value_counts().to_dict()
# to replace values we will use above dict to replace values from column
# here we are doing label encoding
df.replace({'PG': 4, 'Grad': 3, 'HSC': 2, 'SSC': 1, 'Phd': 5},inplace=True)
df
```

## concat()

```python
syntax:
    here operations are performed on column and row names
    by default it performs outer join
    all unique values are present in result
    new_df = pd.concat([df1,df2,...],ignore_index=False,axis = 1,join='outer'/'inner')
    ignore_index-->to ignore_index from original dataframes(True/False) it removes row/column names based on axis
    axis--> 1=column addidition operation 0=row addidition operation
    join--> 'outer'-->it is by default and gives all unique values
            'inner'-->it gives only common index and rows based on axis
```

```python
array = np.random.randint(10,20,size = (3,4))
df1 = pd.DataFrame(array,columns=list('ABCD'))
print(df1)
array = np.random.randint(20,30,size = (3,4))
df2 = pd.DataFrame(array,columns=list('ABCD'))
print(df2)
ndf=pd.concat([df1,df2])
ndf=pd.concat([df1,df2],ignore_index=False)
ndf=pd.concat([df1,df2],ignore_index=True)
ndf=pd.concat([df1,df2],axis=1) #column adding operation 
ndf=pd.concat([df1,df2],axis=0) #row addition operation
ndf
```

## merge()

```python
syntax:
    pd.merge(df1,df2,on=column name which is common,how=inner/outer/left/right,suffixes=['_suffix1','_suffix2'])
    merge works on values of dataframe
    how-->which join we want default is inner
    on-->column on which we compare values
```

```python
df1 = pd.DataFrame({"Name" :['Mangesh','Akash','Sanjay','Sachin','Virat'],
                   'Age':[20,25,27,24,30],
                   'Location':['Pune','Mumbai','Delhi','Chennai','Hyderabad']},
                  index = list('abcde'))

df2 = pd.DataFrame({"Name" :['Mangesh','Swapnil','Akshay','Sachin'],
                   'Age':[20,29,34,24],
                   'Location':['Pune','Kolkata','Chandigarh','Chennai'],
                   'Salary':[30000,40000,60000,70000]},
                  index = list('asdf'))
print(df1)
print(df2)

df3 = pd.DataFrame({"Emp_Name" :['Mangesh','Akash','Sanjay','Sachin','Virat'],
                   'Age':[20,25,27,24,30],
                   'Location':['Pune','Mumbai','Delhi','Chennai','Hyderabad']},
                  index = list('abcde'))

df4 = pd.DataFrame({"Name" :['Mangesh','Swapnil','Akshay','Sachin'],
                   'Age':[20,29,34,24],
                   'Place':['Pune','Kolkata','Chandigarh','Chennai'],
                   'Salary':[30000,40000,60000,70000]},
                  index = list('asdf'))
print(df3)
print(df4)
```

```python
# here default join is inner and will give rows having common values in buth dataframe 
ndf=pd.merge(df1,df2)

# outer join gives all values just like outer join in sql
ndf=pd.merge(df1,df2,how='outer')

# outer join gives all values just like left join in sql
ndf=pd.merge(df1,df2,how='left')

# outer join gives all values just like right join in sql
ndf=pd.merge(df1,df2,how='right')

# on name column 
ndf=pd.merge(df1,df2,on='Name',how='outer')

# on clomun having different names
ndf1=pd.merge(df3,df4,left_on='Emp_Name',right_on='Name',how='outer')

# adding suffixes to column names to identitify column is from which dataframe
# it adds suffixes to column having common names
# default suffixes are x and y
ndf=pd.merge(df1,df2,on='Name',how='outer',suffixes=['_df_one','_df_two'])

ndf1
```
